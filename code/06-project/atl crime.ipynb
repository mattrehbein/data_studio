```python
# race_df = top_ten.groupby('GANG_NAME')['RACE'].value_counts().unstack().reset_index().fillna(0)

atl_solved = atl.groupby('Year')['Solved'].value_counts().unstack().reset_index()
```

```python
ax = atl_solved.plot(kind='bar', x='Year', stacked=True, figsize=(15,7))

ax.grid(False, axis='x')
ax.yaxis.label.set_visible(False)
ax.xaxis.label.set_visible(False)
# ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.5)

plt.show()
```

```python
# percentage of unsolved
atl_unsolved = atl_solved
atl_unsolved['per_unsolved'] = round(atl_solved['No'] / (atl_solved['Yes'] + atl_solved['No']) * 100, 1)
```

```python
atl_unsolved.head() 
```

```python
#C10534', '#ffd200', '#e37e00', '#00887d', '#7ad2f6', '#014d64', '#938DD2'
ax = atl_solved.plot(kind="bar", x='Year', y='per_unsolved', color='#C10534', figsize=(15, 5))

ax.legend().set_visible(False)
ax.grid(False, axis='x')
ax.yaxis.label.set_visible(False)
ax.xaxis.label.set_visible(False)

plt.show()
```
