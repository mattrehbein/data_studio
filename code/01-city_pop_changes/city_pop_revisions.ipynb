```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
TXeco_df = pd.read_excel("UnemploymentReport.xlsx", skiprows=1)
```

```python
my_counties = ['Bexar County, TX', 'Dallas County, TX', 'Tarrant County, TX']
TXco_df = TXeco_df[TXeco_df['Name'].isin(my_counties)]
TXco_df
```

```python
# CHECKING EXACT COL NAMES TO MAKE SURE THEY ARE IDENTICAL BEFORE CONCAT
tx_cols = TXco_df.columns.values.tolist()
tx_cols
```

```python
CAeco_df = pd.read_excel("CAunemploymentReport.xlsx", skiprows=1)
LACo_df = CAeco_df[CAeco_df.Name == "Los Angeles County, CA"]
LACo_df
```

```python
# la_cols = LACo_df.columns.values.tolist()
# la_cols
```

```python
AZeco_df = pd.read_excel("ARUnemploymentReport.xlsx", skiprows=1)
Phoco_df = AZeco_df[AZeco_df.Name == "Maricopa County, AZ"]
Phoco_df
```

```python
# AZ_cols = Phoco_df.columns.values.tolist()
# AZ_cols
```

```python
frames = [TXco_df, LACo_df, Phoco_df]
inc_eco_df = pd.concat(frames, ignore_index=True, sort=False)
inc_eco_df
```

```python
inc_eco_df['City'] = ['San Antonio, Texas', 'Dallas, Texas', 'Fort Worth, Texas', 'Los Angeles, California', 'Phoenix, Arizona']
inc_eco_df
```

```python
pop_df = pd.read_csv("my_pop_df.csv")
pop_df.head()
```

```python
top_pop_df = pop_df.sort_values(by='Pop_Change', ascending=False).head()
top_pop_df
```

```python
merged_df = pd.merge(top_pop_df, inc_eco_df, on='City', how='outer')
merged_df
```

```python
merged_df.to_csv("pop_gain_master.csv", index=False)
```

```python
gain_df = pd.read_csv("pop_gain_master.csv")
```

```python
ax = gain_df.sort_values(by='Pop_Change', ascending=True).plot(kind='barh', color='lightgreen', x='City', y='Pop_Change', figsize=(10,3))

ax.set_title("Cities That Gained The Most People 2016-2017", pad=20)
ax.title.set_fontsize(25)
ax.title.set_fontname("Andale Mono")

ax.tick_params(which='major', left=False, bottom=False, grid_alpha=0.5)

ax.yaxis.label.set_visible(False)

ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)
```
