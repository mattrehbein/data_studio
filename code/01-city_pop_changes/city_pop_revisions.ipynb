```python
import pandas as pd
import matplotlib.pyplot as plt
from cycler import cycler

%matplotlib inline
```

```python
TXeco_df = pd.read_excel("UnemploymentReport.xlsx", skiprows=1)
```

```python
my_counties = ['Bexar County, TX', 'Dallas County, TX', 'Tarrant County, TX']
TXco_df = TXeco_df[TXeco_df['Name'].isin(my_counties)]
TXco_df
```

```python
# CHECKING EXACT COL NAMES TO MAKE SURE THEY ARE IDENTICAL BEFORE CONCAT
tx_cols = TXco_df.columns.values.tolist()
tx_cols
```

```python
CAeco_df = pd.read_excel("CAunemploymentReport.xlsx", skiprows=1)
LACo_df = CAeco_df[CAeco_df.Name == "Los Angeles County, CA"]
LACo_df
```

```python
# la_cols = LACo_df.columns.values.tolist()
# la_cols
```

```python
AZeco_df = pd.read_excel("ARUnemploymentReport.xlsx", skiprows=1)
Phoco_df = AZeco_df[AZeco_df.Name == "Maricopa County, AZ"]
Phoco_df
```

```python
# AZ_cols = Phoco_df.columns.values.tolist()
# AZ_cols
```

```python
frames = [TXco_df, LACo_df, Phoco_df]
inc_eco_df = pd.concat(frames, ignore_index=True, sort=False)
inc_eco_df
```

```python
inc_eco_df['City'] = ['San Antonio, Texas', 'Dallas, Texas', 'Fort Worth, Texas', 'Los Angeles, California', 'Phoenix, Arizona']
inc_eco_df
```

```python
pop_df = pd.read_csv("my_pop_df.csv")
pop_df.head()
```

```python
top_pop_df = pop_df.sort_values(by='Pop_Change', ascending=False).head()
top_pop_df
```

```python
merged_df = pd.merge(top_pop_df, inc_eco_df, on='City', how='outer')
merged_df
```

```python
merged_df.to_csv("pop_gain_master.csv", index=False)
```

```python
loss_df = pd.read_csv("pop_loss_master.csv")
loss_df
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
ax2 = loss_df.sort_values(by='Pop_Change', ascending=False).plot(kind='barh', color=color, x='City', y='Pop_Change', figsize=(15,6))

ax2.set_title("Cities that Lost the Most People 2016-2017", fontweight='bold', pad=20)
ax2.title.set_fontsize(22)
ax2.title.set_fontname("Khmer Sangam MN")

ax2.legend_.remove()
ax2.yaxis.label.set_visible(False)

for i, bar in enumerate(ax2.patches):
    value = "{:,}".format(bar.get_width())
    x_pos = bar.get_width() + 200
    y_pos = i
    ax2.text(x_pos, y_pos, value, va='center', fontweight='bold', size='large', color='white')

plt.savefig("lost_pop.pdf", transparent=True)
```

```python
color = plt.rcParams['axes.prop_cycle'].by_key()['color'][0]
ax3 = loss_df.plot(kind='bar', color=color, x='City', y='2017', width=0.8, figsize=(7,4))

ax3.set_title("Shrinking Cities' 2017 Unemployment Rates", fontweight='bold', pad=15)
ax3.title.set_fontsize(22)
ax3.title.set_fontname("Khmer Sangam MN")

ax3.legend_.remove()
ax3.xaxis.label.set_visible(False)
ax3.yaxis.label.set_visible(False)

ax3.axhline(4.35, color='#7ad2f6')

for i, bar in enumerate(ax3.patches):
    value = str(bar.get_height()) + "%"
    x_pos = i
    y_pos = bar.get_height() - 0.5
    ax3.text(x_pos, y_pos, value, va='top', ha='center', fontweight='bold', size='x-large', color='white')

plt.savefig("lost_unemploy.pdf", transparent=True)
```
